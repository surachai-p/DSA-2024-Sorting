```python
import time

# ฟังก์ชัน Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    
    for i in range(n):
        swapped = False  # ตัวแปรเพื่อตรวจสอบว่ามีการสลับตำแหน่งในรอบนี้หรือไม่
        
        # ลูปสำหรับเปรียบเทียบคู่ตัวเลขที่ติดกัน
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                # ถ้า arr[j] มากกว่า arr[j+1] ให้สลับตำแหน่ง
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True  # ทำการตั้งค่า swapped เป็น True เพราะมีการสลับ
        
        # ถ้าไม่มีการสลับในรอบนี้ แสดงว่าข้อมูลเรียงลำดับแล้ว
        if not swapped:
            break  # หยุดการทำงานทันที เพราะข้อมูลได้เรียงลำดับแล้ว
    
    return arr

# ฟังก์ชัน Insertion Sort
def insertion_sort(arr):
    # วนลูปตั้งแต่ตัวที่ 2 ถึงตัวสุดท้าย
    for i in range(1, len(arr)):
        key = arr[i]  # ค่าที่จะนำไปแทรก
        j = i - 1  # ดัชนีของตัวก่อนหน้า
        
        # ย้ายตัวที่มีค่ามากกว่า key ไปทางขวา
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        
        # แทรก key ลงในตำแหน่งที่เหมาะสม
        arr[j+1] = key
    
    return arr


# ชุดข้อมูลที่ใช้ทดสอบ
test_data = [64, 34, 25, 12, 22, 11, 45, 24, 6, 90]

# ทดสอบฟังก์ชัน Bubble Sort
print("ทดสอบ Bubble Sort:")
print("ข้อมูลก่อนเรียง:", test_data)

start_time = time.time()
sorted_data_bubble = bubble_sort(test_data.copy())
end_time = time.time()

print("ข้อมูลหลังเรียง:", sorted_data_bubble)
print(f"เวลาที่ใช้: {(end_time - start_time)*1000:.6f} มิลลิวินาที\n")


# ทดสอบฟังก์ชัน Insertion Sort
print("ทดสอบ Insertion Sort:")
print("ข้อมูลก่อนเรียง:", test_data)

start_time = time.time()
sorted_data_insertion = insertion_sort(test_data.copy())
end_time = time.time()

print("ข้อมูลหลังเรียง:", sorted_data_insertion)
print(f"เวลาที่ใช้: {(end_time - start_time)*1000:.6f} มิลลิวินาที")
```

```python
def insertion_sort_with_steps(arr):
    print(f"เริ่มต้น: {arr}")
    
    for i in range(1, len(arr)):
        key = arr[i]  # ค่าที่จะนำไปแทรก
        j = i - 1  # ดัชนีของตัวก่อนหน้า
        
        print(f"\nรอบที่ {i}: พิจารณา key = {key}")
        
        # ย้ายตัวที่มีค่ามากกว่า key ไปทางขวา
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
            print(f"  ย้าย {arr[j+1]} ไปทางขวา: {arr}")  # แสดงผลการย้าย
        
        # แทรก key ลงในตำแหน่งที่เหมาะสม
        arr[j+1] = key
        print(f"  แทรก {key} ลงในตำแหน่ง {j+1}: {arr}")  # แสดงผลการแทรก
    
    return arr

# ทดสอบแสดงขั้นตอน
test_data = [64, 34, 25, 12, 22, 11, 45, 24, 6, 90]
insertion_sort_with_steps(test_data.copy())
```
