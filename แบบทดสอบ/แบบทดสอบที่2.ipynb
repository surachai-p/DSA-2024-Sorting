1. ปรับปรุงอัลกอริทึม Insertion Sort ให้รองรับการเรียงลำดับจากมากไปน้อย

```python
def insertion_sort_descending(arr):
    print(f"เริ่มต้น: {arr}")
    
    for i in range(1, len(arr)):
        key = arr[i]  # ค่าที่จะนำไปแทรก
        j = i - 1  # ดัชนีของตัวก่อนหน้า
        
        print(f"\nรอบที่ {i}: พิจารณา key = {key}")
        
        # ย้ายตัวที่มีค่าน้อยกว่า key ไปทางขวา (เพื่อเรียงจากมากไปน้อย)
        while j >= 0 and arr[j] < key:
            arr[j+1] = arr[j]
            j -= 1
            print(f"  ย้าย {arr[j+1]} ไปทางขวา: {arr}")  # แสดงผลการย้าย
        
        # แทรก key ลงในตำแหน่งที่เหมาะสม
        arr[j+1] = key
        print(f"  แทรก {key} ลงในตำแหน่ง {j+1}: {arr}")  # แสดงผลการแทรก
    
    return arr

# ทดสอบแสดงขั้นตอน
test_data = [64, 34, 25, 12, 22, 11, 45, 24, 6, 90]
insertion_sort_descending(test_data.copy())
```

2. ทดสอบกับชุดข้อมูลที่มีค่าซ้ำกัน เช่น `[3, 1, 4, 1, 5, 9, 2, 6, 5, 9]` และตรวจสอบผลลัพธ์

```python
def insertion_sort_descending(arr):
    print(f"เริ่มต้น: {arr}")
    
    for i in range(1, len(arr)):
        key = arr[i]  # ค่าที่จะนำไปแทรก
        j = i - 1  # ดัชนีของตัวก่อนหน้า
        
        print(f"\nรอบที่ {i}: พิจารณา key = {key}")
        
        # ย้ายตัวที่มีค่าน้อยกว่า key ไปทางขวา (เพื่อเรียงจากมากไปน้อย)
        while j >= 0 and arr[j] < key:
            arr[j+1] = arr[j]
            j -= 1
            print(f"  ย้าย {arr[j+1]} ไปทางขวา: {arr}")  # แสดงผลการย้าย
        
        # แทรก key ลงในตำแหน่งที่เหมาะสม
        arr[j+1] = key
        print(f"  แทรก {key} ลงในตำแหน่ง {j+1}: {arr}")  # แสดงผลการแทรก
    
    return arr

# ชุดข้อมูลที่มีค่าซ้ำ
test_data_with_duplicates = [3, 1, 4, 1, 5, 9, 2, 6, 5, 9]

# ทดสอบแสดงขั้นตอน
insertion_sort_descending(test_data_with_duplicates.copy())
```
