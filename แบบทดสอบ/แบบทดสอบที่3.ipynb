1. ปรับปรุงอัลกอริทึม Selection Sort ให้รองรับการเรียงลำดับจากมากไปน้อย

```python
def selection_sort_desc(arr):
    n = len(arr)
    
    for i in range(n):
        max_idx = i  # ตั้งให้ max_idx เริ่มต้นที่ตำแหน่ง i
        print(f"\nรอบที่ {i+1}:")
        print(f"  ข้อมูลปัจจุบัน: {arr}")
        print(f"  ค้นหาค่ามากที่สุดในตำแหน่ง {i} ถึง {n-1}")
        
        for j in range(i+1, n):
            print(f"    เปรียบเทียบ {arr[max_idx]} กับ {arr[j]}", end=" -> ")
            if arr[j] > arr[max_idx]:  # ค้นหาค่ามากที่สุด
                max_idx = j
                print(f"พบค่าที่มากกว่า: {arr[j]}")
            else:
                print("ไม่มีการเปลี่ยนแปลง")
        
        print(f"  ค่ามากที่สุดคือ {arr[max_idx]} ที่ตำแหน่ง {max_idx}")
        if i != max_idx:
            print(f"  สลับ {arr[i]} กับ {arr[max_idx]}")
            arr[i], arr[max_idx] = arr[max_idx], arr[i]
        else:
            print(f"  ไม่ต้องสลับเนื่องจากอยู่ในตำแหน่งที่ถูกต้องแล้ว")
        
        print(f"  ข้อมูลหลังรอบที่ {i+1}: {arr}")
    
    return arr
```

2. วัดประสิทธิภาพเมื่อทดสอบกับชุดข้อมูลที่เกือบเรียงลำดับแล้ว (nearly sorted data)

```python
import time

def selection_sort_desc(arr):
    n = len(arr)
    
    for i in range(n):
        max_idx = i  # ตั้งให้ max_idx เริ่มต้นที่ตำแหน่ง i
        print(f"\nรอบที่ {i+1}:")
        print(f"  ข้อมูลปัจจุบัน: {arr}")
        print(f"  ค้นหาค่ามากที่สุดในตำแหน่ง {i} ถึง {n-1}")
        
        for j in range(i+1, n):
            print(f"    เปรียบเทียบ {arr[max_idx]} กับ {arr[j]}", end=" -> ")
            if arr[j] > arr[max_idx]:  # ค้นหาค่ามากที่สุด
                max_idx = j
                print(f"พบค่าที่มากกว่า: {arr[j]}")
            else:
                print("ไม่มีการเปลี่ยนแปลง")
        
        print(f"  ค่ามากที่สุดคือ {arr[max_idx]} ที่ตำแหน่ง {max_idx}")
        if i != max_idx:
            print(f"  สลับ {arr[i]} กับ {arr[max_idx]}")
            arr[i], arr[max_idx] = arr[max_idx], arr[i]
        else:
            print(f"  ไม่ต้องสลับเนื่องจากอยู่ในตำแหน่งที่ถูกต้องแล้ว")
        
        print(f"  ข้อมูลหลังรอบที่ {i+1}: {arr}")
    
    return arr

# ชุดข้อมูลเกือบเรียงลำดับแล้ว
nearly_sorted_data = [1, 2, 3, 4, 5, 6, 7, 8, 10, 9]

# ทดสอบการทำงานของ Selection Sort (จากมากไปน้อย) บนชุดข้อมูลเกือบเรียงลำดับ
print("ข้อมูลก่อนเรียงลำดับ:", nearly_sorted_data)

start_time = time.time()
sorted_data = selection_sort_desc(nearly_sorted_data.copy())
end_time = time.time()

# ผลลัพธ์หลังจากการเรียงลำดับ
print("\nข้อมูลหลังการเรียงลำดับจากมากไปน้อย:", sorted_data)
print(f"เวลาที่ใช้ในการทดสอบ (มิลลิวินาที): {(end_time - start_time)*1000:.6f} มิลลิวินาที")
```
